# Energy Systems Scenarios

This notebook is a lightweight starting point for renewable generation, demand, storage dispatch, and emissions accounting.

```python
import csv
from pathlib import Path

path = Path('../data/synthetic/hourly_energy_profile.csv')
with path.open() as handle:
    rows = list(csv.DictReader(handle))

total_demand = sum(float(r['demand_mwh']) for r in rows)
total_renewable = sum(float(r['solar_mwh']) + float(r['wind_mwh']) for r in rows)
total_demand, total_renewable, total_renewable / total_demand
```
